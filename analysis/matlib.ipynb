```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
a=np.arange(1,15)
b=a*a*a
c=a
d=np.sin(a)
```

```python
## data scatter
plt.scatter(a,b,c='r')
```

```python
plt.subplot(2,2,1)
plt.scatter(a,b,c='r')
plt.xlabel("X axis")
plt.subplot(2,2,2)
plt.scatter(a,c,c='b',)
plt.xlabel("X axis")
```

```python
## plots
plt.plot(a,b,'r')
```

```python
##creating subplots
plt.subplot(2,2,1)
plt.title("subplots")
plt.plot(a,b,c='r')
plt.subplot(2,2,2)
plt.plot(a,b,'g')
plt.subplot(2,2,3)
plt.plot(a,c,'b')
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.subplot(2,2,4)
plt.plot(a,d,'y--')
plt.xlabel("sin wave")
```

```python
##bar graph

plt.bar(a,b)
```

```python
## histogram
plt.hist(a)
```

```python
## boxplot
plt.boxplot(a,vert=True,patch_artist=True)
```

```python
##pie chart
labels="maths" ,"oops","dsa","se"
sizes=[215,120,180,200]
colors=['gold','green','red','blue']
explode=(0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
plt.axis('equal')
```
